```python
import os
from rdkit import Chem
from config import sdf_file, processed_dir

output_dir = f"{processed_dir}/ligands"
os.makedirs(output_dir, exist_ok=True)

def split_sdf_by_ligands(input_sdf_file, output_directory, ligands_per_file):
    # Create the output directory if it doesn't exist
    os.makedirs(output_directory, exist_ok=True)
    
    # Initialize counters
    current_file_index = 1
    ligand_counter = 0
    current_writer = None
    
    # Read the input SDF file
    suppl = Chem.SDMolSupplier(input_sdf_file)
    
    for mol in suppl:
        if mol is not None:
            # Check if we need to start a new output SDF file
            if ligand_counter % ligands_per_file == 0:
                # Close the current writer if it exists
                if current_writer is not None:
                    current_writer.close()
                
                # Create a new output SDF file
                output_file_path = os.path.join(output_directory, f"output_{current_file_index}.sdf")
                current_writer = Chem.SDWriter(output_file_path)
                current_file_index += 1
            
            # Write the molecule to the current output SDF file
            current_writer.write(mol)
            ligand_counter += 1
    
    # Close the last writer
    if current_writer is not None:
        current_writer.close()

    print(f"Split complete. {ligand_counter} ligands were split into {current_file_index - 1} output files.")

# Specify the input SDF file and output directory
input_sdf_path = '3pages.sdf'
folder = input_sdf_path.replace(".sdf","")


# Specify the desired number of ligands per output SDF file
ligands_per_file = 1



# Call the function to split the SDF file
split_sdf_by_ligands(sdf_file, output_dir, ligands_per_file)
```
